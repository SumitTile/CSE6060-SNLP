# Web Scrapping

```python
import spacy
nlp = spacy.load('en_core_web_sm')

import requests 
url = "http://www.hubertiming.com/results/2017GPTR10K"
#This website conatins data of racing and this is collected in variable
response = requests.get(url) 
```

```python
#checking type of response
type(response) 
```

```python
#Converting the text in response in an html page format
html_page = response.text
```

```python
#Now we are cleaning the text using predefined function beautifulsoup
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_page, 'html.parser')
type(soup)
```

```python
#Cleaning and priting the document
doc = soup.get_text()
print(doc)
```

```python
import nltk
```

```python
#Tokenizing the document containing race data
tokens = nltk.word_tokenize(doc)
```

```python
tokens
```

```python
#Pos tagging the data
pos_tags = nltk.pos_tag(tokens)
```

```python
pos_tags
```

```python
#find the NER in document
doc = nlp(doc)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
