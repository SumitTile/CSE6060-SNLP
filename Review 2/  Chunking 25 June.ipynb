### Chunking

```python
import nltk
```

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
 

grammar = "NP: {<DT>?<JJ>*<NN>}"
```

```python
cp = nltk.RegexpParser(grammar)
```

```python
# Parsing sentence to the regexparser
result = cp.parse(sentence)
```

```python
print(result)
```

```python
result.draw()
```

```python
#Tagging each word in sentence
sentence = """I am pleased to recommend MrXX for an MS in Computer Science at your esteemed university. I have known him since his second year. He was my student in the 3rd semester (2nd year), where I taught him the course of Database Management Systems. """
tken = nltk.word_tokenize(sentence)
tags = nltk.pos_tag(tken)
result = cp.parse(tags)
```

```python
print(result)
```

```python
result.draw()
```

```python
chunks = nltk.ne_chunk(tags)
```

```python
print(chunks)
```

```python
chunks.draw()
```

### NER

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
example = "YouTube is an American online video-sharing platform headquartered in San Bruno, California. Three former PayPal employees—Chad Hurley, Steve Chen, and Jawed Karim—created the service in February 2005. Google bought the site in November 2006 for US$1.65 billion; YouTube now operates as one of Google's subsidiaries.."

doc = nlp(example)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.render(doc, style='ent')
```

```python
displacy.render(doc, style='dep')
```
